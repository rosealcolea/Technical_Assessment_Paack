## Defining the KPIs

After an overall exploration of the data and checking that all the data is in order, I proceed to define the KPIs separating them in 2 groups: general company's performance and individual driver's performance.

**Company:**

- Avg Shift Length
- Order success rate (%)
- Average Time per Order Delivered
- Early Orders vs Late Orders Ratio


**Driver:**

- Avg orders/hour
- Order Success %
- Shift Length metrics
- % of orders delivered compared to other drivers


### Company

#### Avg Shift Length

```python
# In order to compare start time with end time I need to first convert the columns' data type
d["start time"] = pd.to_datetime(d["start time"], format='%H:%M:%S')
d["end time (last delivery attempt)"] = pd.to_datetime(d["end time (last delivery attempt)"], format='%H:%M:%S')
```

```python
# Adding a column that calculates each driver's shift length
d['shift_length'] = d['end time (last delivery attempt)'] - d['start time']
d.head()
```

```python
# Calculating the average of all drivers
avg_shift_length = d['shift_length'].sum()/d['shift_length'].count()
print("Average Shift Length: ", avg_shift_length)
```

> **Observations:** I can observe that some drivers have a significantly inferior shift time compared to their colleagues. I wonder why they would have such short shifts and whether that is a productive strategy. I will check for outliers in this area later to see if it's a rare occurence that can be dealt with individually, or if it's a pattern with a logical explanation.

#### Order Success Rate

```python
d["order_success_rate"] = d['orders delivered']/d['orders attempted']
d.head()
```

```python
# Calculating the average of all drivers
avg_success_ratio = d['order_success_rate'].sum()/d['order_success_rate'].count()
print("Average Success Ratio: ", avg_success_ratio)
```

#### Average Time per Order Delivered

```python
# Creating a function that will transform the time into a numerical (hours)
def hours(x):
    lst = re.findall('[0-9]*', str(x))
    hours = int(lst[7]) + int(lst[9])/60 + int(lst[11])/3600
    return hours

# Creating a new column applying the function just created
d['shift_length_hours'] = [hours(d['shift_length'][i]) for i in range(len(d))]
d.head()
```

```python
d['avg_time_per_order'] = d['shift_length_hours']/d['orders attempted']
d.head()
```

```python
# Calculating the average of all drivers
avg_time_per_order = d['avg_time_per_order'].sum()/d['avg_time_per_order'].count()
print("Average Time per Order: ", avg_time_per_order)
```

> **Observations:** It's interesting to see that this average is so high given that just from the first 5 rows I can see that some driver's take around 5 minutes in average per order. It might be worth checking for outliers in this section to check if only a few drivers take considerably longer per order. 
This could be due to them having a more extense delivery area, other factors, or simply poor performance.

#### Early Orders vs Late Orders Ratio

```python
d['early_orders'] = d['orders early']/d['orders attempted']
d['late_orders'] = d['orders delayed']/d['orders attempted']
d.head()
```

```python
# Calculating the average of all drivers
avg_early_orders_ratio = d['early_orders'].sum()/d['early_orders'].count()
print("Average Early Orders Ratio: ", avg_early_orders_ratio)

avg_late_orders_ratio = d['late_orders'].sum()/d['late_orders'].count()
print("Average Late Orders Ratio: ", avg_late_orders_ratio)
```

> **Observations:** It's interesting to observe that there are significantly more early orders than late orders. This might seem like a sign of high productivity, but an early order is still not being delivered at the requested time, which was the major complaint expressed by the customers.

> This is a major breakthrough that needs to be investigated further and dealt with to mitigate the complaints and further reduce the failed deliveries.

### Driver

#### Avg Orders per Hour

```python
d['avg_orders/hour'] = d['orders attempted']/d['shift_length_hours']
d.head()
```

```python
# Calculating the average of all drivers
avg_orders_per_hour = d['avg_orders/hour'].sum()/d['avg_orders/hour'].count()
print("Average Orders per Hour: ", avg_orders_per_hour)
```

#### Order Sucess Rate

Same calculation we did for the company so simply looking at the column `order success rate` will yiled the information for individual drivers

#### Productivity
Avg of orders delivered per hour worked

```python
d['productivity'] = d['orders delivered']/d['shift_length_hours']
d.head()
```

```python
# Calculating the average of all drivers
avg_productivity = d['productivity'].sum()/d['productivity'].count()
print("Average Productivity: ", avg_productivity)
```

#### Percentage of Orders Delivered Compared to Other Drivers

```python
d['relative_productivity'] = d['productivity']/avg_productivity
d.head()
```

> **Observations:** With this metric we can clearly see who is under performing and who is over performing. We will investigate this metric further later on. It would also be interesting to define a threshold that drivers have to cross in order to be considered productive, and a threshold that would indicate that a given driver's productivity needs to be improved immediately or else we should consider replacing them.

#### Polygon Radar Chart - Single (showing individual driver performance)

```python
# Set data
df = pd.DataFrame({
'group': ['A','B','C','D'],
'Order Success Rate': [38, 1.5, 30, 4],
'Early Orders Rate': [29, 10, 9, 34],
'Late Orders Rate': [8, 39, 23, 24],
'Relative Productivity per Order': [7, 31, 33, 14],
'Relative Productivity per Parcel': [28, 15, 32, 14]
})
 
# number of variable
categories=list(df)[1:]
N = len(categories)
 
# We are going to plot the first line of the data frame.
# But we need to repeat the first value to close the circular graph:
values=df.loc[0].drop('group').values.flatten().tolist()
values += values[:1]
values
 
# What will be the angle of each axis in the plot? (we divide the plot / number of variable)
angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]
 
# Initialise the spider plot
ax = plt.subplot(111, polar=True)
 
# Draw one axe per variable + add labels labels yet
plt.xticks(angles[:-1], categories, color='grey', size=8)
 
# Draw ylabels
ax.set_rlabel_position(0)
plt.yticks([10,20,30], ["10","20","30"], color="grey", size=7)
plt.ylim(0,40)
 
# Plot data
ax.plot(angles, values, linewidth=1, linestyle='solid')
 
# Fill area
ax.fill(angles, values, 'b', alpha=0.1)
```

#### Polygon Radar Chart - Double (comparing 2 drivers' performance)

```python
# Set data
df = pd.DataFrame({
'group': ['A','B','C','D'],
'Order Success Rate': [38, 1.5, 30, 4],
'Early Orders Rate': [29, 10, 9, 34],
'Late Orders Rate': [8, 39, 23, 24],
'Relative Productivity per Order': [7, 31, 33, 14],
'Relative Productivity per Parcel': [28, 15, 32, 14]
})
 
 
 
# ------- PART 1: Create background
 
# number of variable
categories=list(df)[1:]
N = len(categories)
 
# What will be the angle of each axis in the plot? (we divide the plot / number of variable)
angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]
 
# Initialise the spider plot
ax = plt.subplot(111, polar=True)
 
# If you want the first axis to be on top:
ax.set_theta_offset(pi / 2)
ax.set_theta_direction(-1)
 
# Draw one axe per variable + add labels labels yet
plt.xticks(angles[:-1], categories)
 
# Draw ylabels
ax.set_rlabel_position(0)
plt.yticks([10,20,30], ["10","20","30"], color="grey", size=7)
plt.ylim(0,40)
 
 
# ------- PART 2: Add plots
 
# Plot each individual = each line of the data
# I don't do a loop, because plotting more than 3 groups makes the chart unreadable
 
# Ind1
values=df.loc[0].drop('group').values.flatten().tolist()
values += values[:1]
ax.plot(angles, values, linewidth=1, linestyle='solid', label="Driver A")
ax.fill(angles, values, 'b', alpha=0.1)
 
# Ind2
values=df.loc[1].drop('group').values.flatten().tolist()
values += values[:1]
ax.plot(angles, values, linewidth=1, linestyle='solid', label="Driver B")
ax.fill(angles, values, 'r', alpha=0.1)
 
# Add legend
plt.legend(loc='lower right', bbox_to_anchor=(0.1, 0.1))
```
